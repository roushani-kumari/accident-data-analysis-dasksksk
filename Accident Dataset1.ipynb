```python
# import Dask
import dask.dataframe as dd

# read CSV file
df = dd.read_csv("Accident Dataset.csv")

# show all Column
print("Column naam:")
print(df.columns)
print()

# Total records kitne hain (rows count)
print("Total records:", df.shape[0].compute())
print()

# Har din kitne accident hue (Monday, Tuesday, etc.)
print("Har din ke accident:")
print(df["Day_of_week"].value_counts().compute())
print()

# Driver ki age group ke hisaab se analysis
print("Driver ki Age ke hisaab se accidents:")
print(df["Age_band_of_driver"].value_counts().compute())
print()

# Driving experience ke hisaab se analysis
print("Driver ka anubhav (experience):")
print(df["Driving_experience"].value_counts().compute())
print()

# Accident kitne severe the (minor, serious, fatal)
print("Accident severity:")
print(df["Accident_severity"].value_counts().compute())
print()

# Kon se vehicle type ka accident zyada hua
print("Vehicle type wise accidents:")
print(df["Type_of_vehicle"].value_counts().compute().head(10))
print()

# Road surface ka type (kon se road pe zyada accident)
print("Road surface type:")
print(df["Road_surface_type"].value_counts().compute())
print()

# Mausam (weather) ke time ka condition
print("Mausam (Weather conditions):")
print(df["Weather_conditions"].value_counts().compute())
print()

# Light condition (andhera, roshni, etc.)
print("Light conditions:")
print(df["Light_conditions"].value_counts().compute())
print()

# Har ghante ke hisaab se accident
print("✅ Har ghante me kitne accident hue:")
print(df["Hour"].value_counts().compute().sort_index())
print()

# Casualty (ghayal hone ka level)
print("Casualty severity:")
print(df["Casualty_severity"].value_counts().compute())
print()

# Accident hone ka reason (kaaran)
print("Accident ka reason:")
print(df["Cause_of_accident"].value_counts().compute().head(10))
```
